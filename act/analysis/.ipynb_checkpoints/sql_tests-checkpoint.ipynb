```python
%%sql

# find the votes won by a specific candidate in all booths of an ac

show tables;
select fbc.contest_booth_no, fab.booth_name, sum(fbc.vote_count)
from fact_booth_vote_count fbc join fact_election_candidate fec
on fbc.candidate_id = fec.candidate_id
join fact_election fe on 
fe.election_id = fbc.election_id
join fact_ac_booth fab on 
fab.booth_no = fbc.contest_booth_no
where fe.election_state = 'Bihar'
and fe.election_year = 2015
and fe.election_type = 'Assembly'
and fec.party_abbr = 'I-1'
and fab.delim_year < fe.election_year

group by 1, 2

order by 3 desc;
```

```python
%%sql

# find turnout ratio of each live booth in an election in an ac

select 
cv.*  
, count(distinct fee.epic) as electors
from fact_electorate fee 
join (select 
        fbc.election_id
        , fe.election_state
        , fbc.contest_pc_no
        , fbc.contest_ac_no
        , fbc.contest_booth_no
        , fab.booth_name
        , sum(vote_count) as votes
    from fact_booth_vote_count fbc join fact_election fe 
    on fbc.election_id = fe.election_id
    join fact_ac_booth fab
    on fab.state = fe.election_state 
    and fab.pc_no = fbc.contest_pc_no 
    and fab.ac_no = fbc.contest_ac_no 
    and fab.booth_no = fbc.contest_booth_no

    where fe.election_state = 'Bihar'
    and fe.election_year = 2015
    and fe.election_type = 'Assembly'
    and fab.delim_year < fe.election_year
    group by 1, 2, 3, 4, 5, 6
) as cv
on fee.state = cv.election_state 
and fee.pc_no = cv.contest_pc_no 
and fee.ac_no = cv.contest_ac_no 
and fee.booth_no = cv.contest_booth_no

group by 1, 2, 3, 4, 5, 6, 7
```

```python
%%sql
select surname, count(distinct epic) as strength from fact_electorate group by surname;
```

```python
def all_replace(name):
    mapping = {"'":'', '(':'', '॥':'', '*':'', ')':'', '-':'', '.':'', ';':'', '[':'', ']':'', '?':'', '!':'', '/':'',
               '1':'', '2':'', '3':'', '4':'', '5':'', '6':'', '7':'', '8':'', '9':'', '0':'',
               '१':'', '२':'', '३':'', '४':'', '५':'', '६':'', '७':'', '८':'', '९':'', '०':'',
               u'\u200d':'', u'\u200c':''
              }
    for k in mapping:
        name = name.replace(k, mapping[k])
    return name
    
res = %sql select surname, count(distinct epic) as strength from fact_electorate group by surname;
df = res.DataFrame()

df['surname_clean'] = df['surname'].map(all_replace, na_action='ignore')

df.to_csv(sep=',', path_or_buf='surnames.csv', header=True, index=False, na_rep="")
```
